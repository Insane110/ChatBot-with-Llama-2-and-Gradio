#Loading Model & Tokenizer

````python
from transformers import AutoTokenizer
import transformers
import torch
````

````python
model = "meta-llama/Llama-2-7b-chat-hf"
tokenizer = AutoTokenizer.from_pretrained(model)
````

#Pipeline

````python
from transformers import pipeline
llama_pipeline = pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    tokenizer=tokenizer,
    device_map="auto",
)
````

The right structure of Llama 2 prompts:

```
<s>[INST] <<SYS>>
{{ system_prompt }}
<</SYS>>

{{ user_message }} [/INST]
```

````python
PROMPT = """<s>[INST] <<SYS>>
You are a really good and helpful bot. Your answers are really intellectual,humourous,clear and concise.
<</SYS>>

"""
````

````python
# Formatting function for message and history
def format(message: str, history: list, memory_limit: int = 5) -> str:
    """
    Formats the message and history for the Llama model.
    Parameters:
        message (str): Current message to send.
        history (list): Past conversation history.
        memory_limit (int): Limit on how many past interactions to consider.
    Returns:
        str: Formatted message string
    """
    # always keep len(history) <= memory_limit
    if len(history) > memory_limit:
        history = history[-memory_limit:] #keeps the most recent memory element

    if len(history) == 0:
        return PROMPT + f"{message} [/INST]"

    formatted_message = PROMPT + f"{history[0][0]} [/INST] {history[0][1]} </s>"

    # Handle conversation history
    for user_msg, model_answer in history[1:]:
        formatted_message += f"<s>[INST] {user_msg} [/INST] {model_answer} </s>"

    # Handle the current message
    formatted_message += f"<s>[INST] {message} [/INST]"

    return formatted_message
````

#Generating Responses

````python
def response(message: str, history: list) -> str:
    """
    Generates a conversational response from the Llama model.
    Parameters:
        message (str): User's input message.
        history (list): Past conversation history.
    Returns:
        str: Generated response from the Llama model.
    """
    query = format(message, history)
    res = ""
    sequences = llama_pipeline(
        query,
        do_sample=True,
        top_k=10,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_length=512,
    )
    generated_response = sequences[0]['generated_text']
    res = generated_response[len(query):]  # Remove the prompt from the output

    print("Chatbot:", res.strip())
    return res.strip()
````

#Launch Chatbot with Gradio


````python
import gradio as gr
gr.ChatInterface(response).launch()
````
